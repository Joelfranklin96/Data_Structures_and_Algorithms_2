```python
#Return the index of first occurence of number
```

```python
def firstIndex_1(arr, x):
    
    if arr[0] == x:
        return 0
    
    if len(arr) == 1:
        return -1
    
    if firstIndex_1(arr[1:],x) == -1:
        return -1
    else:
        return firstIndex_1(arr[1:],x) + 1
    
# It's a good practice to replace firstIndex_1(arr[1:],x) = smaller_output so that we don't need to calculate
# firstIndex_1(arr[1:],x) over and over again.
```

```python
print(firstIndex_1([1,3,4,5,5,8],5))
print(firstIndex_1([1,3,4,5,5,8],6))
print(firstIndex_1([1,3,4,5,5,8],8))
print(firstIndex_1([1,3,4,5,6,8,6,3],6))
```

```python
def firstIndex_2(arr,x,si):
    
    l = len(arr)
    
    if arr[si] == x:
        return si
    
    if si == l-1:
        return -1
    
    return firstIndex_2(arr,x,si+1)
```

```python
print(firstIndex_2([1,3,4,5,5,8],5,0))
print(firstIndex_2([1,3,4,5,5,8],6,0))
print(firstIndex_2([1,3,4,5,5,8],8,0))
print(firstIndex_2([1,3,4,5,6,8,6,3],6,0))
```
