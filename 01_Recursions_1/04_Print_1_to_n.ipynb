```python
#Print numbers 1 to n using Recursion
```

```python
def f(b):
    
    if b == 1:
        print(b)
    else:
        f(b-1)
        print(b)
```

```python
f(5)
```

```python
#Print numbers n to 1 using Recursion
```

```python
def f(b):
    
    if b == 1:
        print(b)
    else:
        print(b)
        f(b-1)
```

```python
f(5)
```
