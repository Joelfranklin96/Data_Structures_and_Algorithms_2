```python
#Return the index of last occurence of number
```

```python
def LastIndex_1(arr,x,si):
    
    if arr[si] == x:
        return si
    
    if si == 0:
        return -1
    
    return LastIndex_1(arr,x,si-1)
```

```python
print(LastIndex_1([1,3,4,5,5,8],5,5))
print(LastIndex_1([1,3,4,5,5,8],6,5))
print(LastIndex_1([1,3,4,5,5,8],8,5))
print(LastIndex_1([1,3,4,5,6,8,6,3],6,7))
```

```python
def LastIndex_2(arr,x):
    
    if len(arr) == 0:
        return -1
    
    if LastIndex_2(arr[1:],x) != -1:
        return LastIndex_2(arr[1:],x) + 1
    else:
        if arr[0] == x:
            return 0
        else:
            return -1
```

```python
print(LastIndex_2([1,3,4,5,5,8],5))
print(LastIndex_2([1,3,4,5,5,8],6))
print(LastIndex_2([1,3,4,5,5,8],8))
print(LastIndex_2([1,3,4,5,6,8,6,3],6))
```
