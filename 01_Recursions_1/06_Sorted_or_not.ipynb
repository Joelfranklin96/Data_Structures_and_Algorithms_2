```python
#Find whether the list is Sorted or not using Recursion
```

```python
def isSorted_1(a):
    
    if len(a) == 0 or len(a) == 1:
        return True
    else:
        if (a[0] > a[1]):
            return False
        return isSorted_1(a[1:])
```

```python
print(isSorted_1([2,3,4,5,8,9]))
print(isSorted_1([2,3,14,5,8,9]))
```

```python
def isSorted_2(a):
    
    if len(a) == 0 or len(a) == 1:
        return True
    else:
        c = a[0] < a[1]
        d = isSorted_2(a[1:])
        return c and d
```

```python
print(isSorted_2([2,3,4,5,8,9]))
print(isSorted_2([2,3,4,5,8,9,3,13,14]))
```

```python
def isSorted_3(a,si):
    
    l = len(a)
    
    if si == l-1:
        return True
    
    if a[si] > a[si + 1]:
        return False
    
    return isSorted_3(a,si+1)
```

```python
print(isSorted_3([2,3,4,5,8,9],0))
print(isSorted_3([2,3,4,5,8,9,3,13,14],0))
print(isSorted_3([2,3,4,5,8,9],0))
print(isSorted_3([2,3,14,5,8,9],0))
```
