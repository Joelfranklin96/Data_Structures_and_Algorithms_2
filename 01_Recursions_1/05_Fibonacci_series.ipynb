```python
#Find Fibonacci series with Recursion
```

```python
def fibonacci_recursion(n):
    

    if n == 1:
        return 1
    elif n == 2:
        return 1
    
    else:
        return fibonacci_recursion(n-1) + fibonacci_recursion(n-2)
```

```python
print(fibonacci_recursion(1))
print(fibonacci_recursion(2))
print(fibonacci_recursion(3))
print(fibonacci_recursion(4))
print(fibonacci_recursion(5))
print(fibonacci_recursion(6))
```

```python
#Find Fibonacci series without Recursion
```

```python
def fibonacci(n):
    
    a = [1,1,2]
    
    if len(a) > n:
        return a[n-1]
    
    else:
    
        while (len(a) < n):
            a.append(a[-1] + a[-2])

        return a[n-1]
```

```python
print(fibonacci(1))
print(fibonacci(2))
print(fibonacci(3))
print(fibonacci(4))
print(fibonacci(5))
print(fibonacci(6))
```
