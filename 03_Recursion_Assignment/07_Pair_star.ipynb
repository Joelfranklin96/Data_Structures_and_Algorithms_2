```python
# Given a string S, compute recursively a new string 
# where identical chars that are adjacent in the original string are separated from each other by a "*".

# Input - "apple"
# Output - "ap*ple"
```

```python
def pair_star(s):
    
    if len(s) == 1:
        return s
    
    smallOutput = pair_star(s[1:])
    
    if s[0] == s[1]:
        return s[0] + "*" + smallOutput
    else:
        return s[0] + smallOutput
```

```python
s = "apple"
pair_star(s)
```
