```python
# Binary Search Algorithm
```

```python
def binary_search(arr,lo,hi,x):
    
    mid = (lo + hi)//2
    
    if hi < lo:
        return "Not Found"
    
    elif arr[mid] == x:
        return "Found"
    
    elif x < arr[mid]:
        hi = mid - 1
        return binary_search(arr,lo,hi,x)
    
    else:
        lo = mid + 1
        return binary_search(arr,lo,hi,x)
```

```python
print(binary_search([1,2,4,5,7,8],0,5,0))
print(binary_search([1,2,4,5,7,8],0,5,7))
print(binary_search([1,2,4,5,7,8],0,5,1))
```
