```python
#Replace character "a" with "e" in the given string
```

```python
def replace(a):
    
    if len(a) == 0:
        return a
    
    smallOutput = replace(a[1:])
    if a[0] =="a":
        return "e" + smallOutput
    else:
        return a[0] + smallOutput
```

```python
print(replace("Jals"))
print(replace("Clap"))
print(replace("mlape"))
```
