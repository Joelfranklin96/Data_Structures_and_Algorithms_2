```python
def tower_of_hanoi(n, a, b, c):
    
    if n == 0:
        return
    
    if n == 1:
        print("Move 1st disk from {} to {}".format(a,c))
        return
    
    tower_of_hanoi(n-1, a, c, b)
    
    print("Move {}th disk from {} to {}".format(n,a,c))
    
    tower_of_hanoi(n-1, b, a, c)
```

```python
tower_of_hanoi(4,"a","b","c")
```
