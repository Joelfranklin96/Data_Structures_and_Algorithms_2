```python
def merge_2(a,b):
    
    merged = []
    
    i = 0
    j = 0
    k = 0
    
    while i < len(a) and j < len(b):
        
        if a[i] <= b[j]:
            
            merged.append(a[i])
            i = i + 1
        
        else:
            
            merged.append(b[j])
            j = j + 1
        
        k = k + 1
        
    a_rem = a[i:]
    b_rem = b[j:]
    
    return merged + a_rem + b_rem
```

```python
def merge_sort_2(arr):
    
    mid = len(arr) // 2
    
    if len(arr) <= 1:
        return arr
    
    left_sorted = merge_sort_2(arr[:mid])
    right_sorted = merge_sort_2(arr[mid:])
    
    whole_sorted = merge_2(left_sorted, right_sorted)
    
    return whole_sorted
```
