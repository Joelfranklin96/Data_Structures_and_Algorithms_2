```python
def pair_sum(arr, n, num) :

    flag = 0
    
    for x in range(n):
        
        for y in range(x+1,n):
            
            if (arr[x] + arr[y] == num):
                flag = flag + 1
    
    return flag
```

```python
pair_sum([3,2,1,3,6,8,4,10],8,5)
```

```python
# Time complexity = O(n^2)
# Space complexity = O(1)
```

```python
def pair_sum_1(arr, n, num):
    
    flag = 0
    
    arr.sort()
    
    for x in range(n):
        
            y = num - arr[x]
            
            if y >= 0:

                flag = flag + arr[x+1:].count(y)
    
    return flag
```
